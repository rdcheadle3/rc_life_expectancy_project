### Create and visualize subsets of data in order to form hypotheses

#### Create subset for Low co2 producing countries

```python
subset_low_co2 = df.loc[df['co2_production_flag'] == 'Low co2 Producer']
```

```python
subset_low_co2.head()
```

#### Correlation matrix of subset_low_co2

```python
# create subset of variables necessary for exploratory visual analysis; remove 'country' and 'year'

sub_low_co2 = subset_low_co2.drop(columns=['country', 'year'])

sub_low_co2.head()
```

```python
# Create a correlation matrix of subset

sub_low_co2.corr()
```

```python
# use the np.triu() numpy function to isolate the upper triangle 
# of a matrix while turning all the values in the lower triangle into 0.

np.triu(np.ones_like(sub_low_co2.corr()))
```

```python
# Create triangle correlation heat map of sub subset

plt.figure(figsize=(16, 6))
# define the mask to set the values in the upper triangle to True
mask = np.triu(np.ones_like(sub_low_co2.corr(), dtype=np.bool))
heatmap = sns.heatmap(sub_low_co2.corr(), mask=mask, vmin=-1, vmax=1, annot=True, cmap='RdBu')
heatmap.set_title('Triangle Correlation Heatmap: Low co2 Producing Countries', fontdict={'fontsize':18}, pad=16);
```

#### Scatter plot : low_co2 subset

```python
# Create a scatter plot comparing co2 production and life expectancy in low co2 producing countries


sns.scatterplot(data=subset_low_co2, x="life_expectancy", y="co2")
```

It is hard to tell if any correlation exists between a countries overall co2 production and the life expectancy of a person in that country in countries considered low co2 producers. I believe we will need to look at co2 production on a per capita basis.
