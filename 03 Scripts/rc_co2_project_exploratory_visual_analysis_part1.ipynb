## 03 Exploratory visual analysis

### Create a correlation matrix heatmap

```python
# create subset of variables necessary for exploratory visual analysis; remove 'country' and 'year'

sub = df.drop(columns=['country', 'year'])

sub.head()
```

```python
# Create a correlation matrix of subset

sub.corr()
```

```python
# use the np.triu() numpy function to isolate the upper triangle 
# of a matrix while turning all the values in the lower triangle into 0.

np.triu(np.ones_like(sub.corr()))
```

```python
# Create triangle correlation heat map of sub subset

plt.figure(figsize=(16, 6))
# define the mask to set the values in the upper triangle to True
mask = np.triu(np.ones_like(sub.corr(), dtype=np.bool))
heatmap = sns.heatmap(sub.corr(), mask=mask, vmin=-1, vmax=1, annot=True, cmap='RdBu')
heatmap.set_title('Triangle Correlation Heatmap', fontdict={'fontsize':18}, pad=16);
```


Correlation Coefficient (r)	Description

+0.8 to +1.0 Very strong + association

+0.6 to +0.8 Strong + association

+0.4 to +0.6 Moderate + association

+0.2 to +0.4 Weak + association

0.0 to +0.2	 Very weak + or no association

0.0 to -0.2	 Very weak - or no association

-0.2 to 0.4 Weak - association

-0.4 to -0.6 Moderate - association

-0.6 to -0.8 Strong - association

-0.8 to -1.0 Very strong - association








```python
# save triangle correlation heat map

heatmap.figure.savefig(os.path.join(path, 'corr_heatmap.png'))
```
