## 03 Import data

```python
# import carbon_life_full csv as df

df = pd.read_csv(os.path.join(path, 'carbon_life_full.csv'), index_col = False)

df.head(5)
```

```python
# drop 'Column1' from df

df = df.drop('Column1', axis=1)
```

```python
df.columns
```

## 05 Derive Per Capita Variables

```python
# create column 'gdp per capita', 'co2 per capita'
# co2 per capita will be in tons not million tons

df['gdp_per_capita'] = df['gdp']/df['population']

df['co2_per_capita'] = (df['co2']*1000000)/df['population']

df.head(5)
```

```python
# change 'co2' from million tons to tons by multiplying values in 'co2' by 1,000,000

df['co2'] = df['co2']*1000000

df.head(5)
```

## 06 Derive % Change Variables

```python
# create column 'pct_change_pop', 'pct_change_gdp', 'pct_chnage_co2', 'pct_change_life'

df['pct_change_pop'] = df['population'].pct_change()

df['pct_change_gdp'] = df['gdp'].pct_change()

df['pct_chnage_co2'] = df['co2'].pct_change()

df['pct_change_life'] = df['life_expectancy'].pct_change()

df.head(5)
```

```python
# create column 'pct_change_gdp_capita'

df['pct_change_gdp_capita'] = df['gdp_per_capita'].pct_change()

df.head(5)
```

```python
# create column 'pct_change_co2_capita'

df['pct_change_co2_capita'] = df['co2_per_capita'].pct_change()

df.head(5)
```

```python
# rename 'pct_chnage_co2' to 'pct_change_co2'

df.rename(columns = {'pct_chnage_co2':'pct_change_co2'}, inplace = True)
```

```python
df = df[['country', 'year', 'life_expectancy', 'pct_change_life', 'population', 'pct_change_pop', 'gdp', 'pct_change_gdp', 'gdp_per_capita', 'pct_change_gdp_capita', 'co2', 'pct_change_co2', 'co2_per_capita', 'pct_change_co2_capita']]

df.head(5)
```

```python
# multiply 'pct_change_pop', 'pct_change_gdp', 'pct_change_gdp_capita', ''pct_change_co2', 'pct_change_co2_capita', and 'pct_change_life' by 100

df['pct_change_pop'] = df['pct_change_pop']*100

df['pct_change_gdp'] = df['pct_change_gdp']*100

df['pct_change_gdp_capita'] = df['pct_change_gdp_capita']*100

df['pct_change_co2'] = df['pct_change_co2']*100

df['pct_change_co2_capita'] = df['pct_change_co2_capita']*100

df['pct_change_life'] = df['pct_change_life']*100

df.head(5)
```

## 07 Export dataframe 

```python
# export df as carbon_life_full_v02.csv

df.to_csv(os.path.join(path, 'carbon_life_full_v02.csv'))
```
